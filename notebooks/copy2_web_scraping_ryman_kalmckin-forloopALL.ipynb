```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

```python
URL = 'https://ryman.com/events/list/?tribe_event_display=list&tribe_paged'
```

```python
multi_titles = []
for page in range(0,5):
    req = requests.get(URL + str(page) + '=')
    soup = BS(req.text, 'html.parser')
    all_events = soup.find_all('a', class_="tribe-event-url")
     
    for i in range(0,19):
        if page>1:
            print(f"{(i+1)+page*20}" + all_events[i].text)
        else:
            print(f"{i+1}" + all_events[i].text)
            
multi_titles.append(all_events)


    
    
    
```

```python
multi_times = []
for page in range(0,5):
    req = requests.get(URL + str(page) + '=')
    soup = BS(req.text, 'html.parser')
    time = soup.find_all('time')
     
    for i in range(0,19):
        if page>1:
            print(f"{(i+1)+page*20}" + time[i].text)
        else:
            print(f"{i+1}" + time[i].text)
            
multi_times.append(time)
```
